# Visualizations

```python
plt.figure(1, figsize=(15,5))
plt.show()

n=0
for x in ['Age','Annual Income (k$)','Spending Score (1-100)']:
    n += 1
    plt.subplot(1,3,n)
    plt.subplots_adjust(hspace=0.5,wspace=0.5)
    sns.distplot(df[x],bins=15)
    plt.title('DP {}'.format(x),rotation=45,size=)

plt.show()
```

```python
sns.pairplot(df, vars=['Annual Income (k$)','Spending Score (1-100)','Age',],hue='Gender')
plt.show()
```

```python
x1 = df[['Age','Spending Score (1-100)']].iloc[:, :].values
inertia = []
for n in range(1,15):
    algorithm = KMeans(
        n_clusters=n,
        init='k-means++',
        n_init=10,
        max_iter=300,
        tol=0.0001,
        random_state=111,
        algorithm='elkan'
    )
    algorithm.fit(x1)
    inertia.append(algorithm.inertia_)
```

```python
plt.figure(1, figsize=(9,4))
plt.plot(np.arange(1,15),inertia,'o')
plt.plot(np.arange(1,15),inertia,'-',alpha=0.5)
plt.xlabel('No. of Clusters'),plt.ylabel('Inertia')
plt.show()
```

```python
algorithm = KMeans(
    n_clusters=4,
    init='k-means++',
    n_init=10,
    max_iter=300,
    tol=0.0001,
    random_state=111,
    algorithm='elkan'
)
algorithm.fit(x1)
labels1=algorithm.labels_
centroids1=algorithm.cluster_centers_

plt.figure(1, figsize=(9,4))
plt.scatter(x='Age',y='Spending Score (1-100)', data=df, c=labels1,s=100)
plt.scatter(x=centroids1[:,0],y=centroids1[:,1], c='red',s=100,alpha=0.5)
plt.xlabel('Age'),plt.ylabel('Spending Score (1-100)')
plt.show()
```
